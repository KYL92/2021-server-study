```python
import cv2
import numpy as np
import onnxruntime
from PIL import Image
```

```python
img = Image.open("test.JPG")
img_resize = img.resize((28, 28))
imgArray = np.asarray(img_resize, dtype=np.float32)
imgArray = imgArray.mean(axis=-1)
```

```python
imgArray = imgArray/255.0
imgArray = imgArray.reshape(1, 28*28)
```

```python

session = onnxruntime.InferenceSession('mnist.onnx', None)
result = session.run(None, {'input': imgArray})[0]
exp_a = np.exp(result[0])
sum_exp_a = np.sum(exp_a)
y = exp_a / sum_exp_a
y
```
